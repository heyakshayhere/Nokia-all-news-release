```python
#importing dependencies
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
#driver for operation
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
%%time

#there are 142 pages
titles = []
dates = []
links = []

#try except loop so that our script will run without errors 
try:
    for page_number in range(0,143):
        #fetching the pages 
        url = f'https://www.nokia.com/about-us/newsroom/press-and-stock-exchange-releases/?page={page_number}/'

        #getting the url
        driver.get(url)

        if page_number == 0:
            #accepting the cookies
            driver.find_element_by_xpath('''//*[@id="modalAcceptAllBtn"]''').click()

        for x in range(1,10):
            title = driver.find_element_by_xpath(f'''//*[@id="press-releases-wrapper"]/div/div[{x}]/div/div/div[1]/div/a/h3/span''').text
            link = driver.find_element_by_xpath(f'''//*[@id="press-releases-wrapper"]/div/div[{x}]/div/div/div[1]/div/a''').get_attribute('href')
            date = driver.find_element_by_xpath(f'''//*[@id="press-releases-wrapper"]/div/div[{x}]/div/div/div[1]/div/label''').text
            titles.append(title)
            dates.append(date)
#             print(f"{title}\n{date}\n{link}\n")
            links.append(link)
        
        print(f"scrapping done for page {page_number}\n")

except:
    pass
```

```python
#importing pandas for dataframe operations
import pandas as pd ,numpy as np 
```

```python
#zippling the list so that it'll form the tuple
zipped = list(zip(dates,titles,links))
zipped[:5]
```

```python
#creating the csv 
csv_file = pd.DataFrame(zipped,columns=['date','title','link'])
csv_file
```

```python
#saving the csv and keeping the index as false
csv_file.to_csv('Nokia all news release link.csv',index=False)
```
