```python
#importing dependencies
import pandas as pd , numpy as np 
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
#importing the data links and txt names
news_urls = pd.read_csv(r"Nokia all news release link.csv")
news_urls
```

```python
news_urls['link']
```

```python
#driver for operation
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
%%time 

titles = []
articles =[]

for link_number in range(len(news_urls)):
    
    #getting the url from data
    url = news_urls['link'][link_number]
    
    #accessing the url
    driver.get(url)
    
    if link_number == 0:
        #accepting the cookies
        driver.find_element_by_xpath('''//*[@id="modalAcceptAllBtn"]''').click()
    
    title = driver.find_element_by_xpath('''/html/body/div[1]/div/main/div[1]/article/div/div/div/div/h1/span''').text
    titles.append(title)
        
    article = driver.find_element_by_xpath('''/html/body/div[1]/div/main/div[2]/div/div/div/div[2]''').text
    articles.append(article)
    
    print(f"scrapping done for page {link_number}\n")   
```

```python
#zippling the list so that it'll form the tuple
zipped = list(zip(titles,articles))
zipped[:5]
```

```python
#creat
csv_file = pd.DataFrame(zipped,columns=['title','article'])
csv_file
```

```python
news_urls['article'] = csv_file['article']
```
